```python
### Scenario 1:
#### User Authentication for BUEA GENERAL HOSPITAL

BUEA GENERAL HOSPITAL is upgrading its internal systems to enhance data security and restrict access to sensitive patient
information. You have been tasked with designing a user authentication system using a `while` loop in Python. Consider the following:

> User Database:

Create a hypothetical user database with at least five user accounts. Each account should have a unique username and password. These accounts represent hospital staff who require access to patient records and other confidential information.

> User Authentication Process:

Implement a Python program using a while loop that simulates the user authentication process. Prompt the user to enter their username and password.
If the entered username is not in the database, print an error message and allow the user to retry.
If the username is correct, prompt for the password. If the password is incorrect, print an error message and allow the user to retry.
If both username and password are correct, print a welcome message and exit the while loop.
```

```python

user_database = {
    "John": "09877",
    "Doe": "yamelexcel",
    "Micheal": "Micheal",
    "Salah": "Salahl237",
    "Cathy": "cathy124"
}


print("BUEA GENERAL HOSPITAL LOGIN SYSTEM ")

while True:
    username = input(" enter username: ")

    if username not in user_database:
        print(" Username not found. Enter a Valid username.\n")
       
        continue
    password = input("Enter password: ")

   
    if user_database[username] != password:
        print("Incorrect password. Please try again.\n")
        continue
    print(f"\n Access Granted!")
    print(f"Welcome {username}, you now have access to patient records.")
    break
```
