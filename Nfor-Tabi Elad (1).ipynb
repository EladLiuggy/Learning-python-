```python
#Question Two
#Njieforbi Restarant inventory management system

# Collect input
item_name = input("Enter item name: ")
current_stock = int(input("Enter current stock level: "))

# Set threshold
threshold = 15

if current_stock < threshold:
    reorder_quantity = threshold - current_stock
    print("Item Name:", item_name)
    print("Current Stock Level:", current_stock)
    print("Status: WARNING - Stock is low")
    print("Suggested Reorder Quantity:", reorder_quantity)
else:
    print("Item Name:", item_name)
    print("Current Stock Level:", current_stock)
    print("Status: Stock is sufficient")
```

```python
#Question one
#Retail shop Record Tracker
#variable Creation

daily_sales = []

#user Entery of values 
for i in range(5):
    sale = int(input("Enter daily sales: "))
    daily_sales.append(sale)

#Calculating total sales for the week 
total_sales = 0
for sale in daily_sales:
    total_sales += sale

# Calculate average sales for the week
average_sales = total_sales / 5

# Weekly expenses of the shop (considering spends 150000 a week)
weekly_cost = 150000

#Determining profit or loss
if total_sales > weekly_cost:
    print("The shop made a PROFIT")
else:
    print("The shop made a LOSS")

# Display results
print("Total Sales:", total_sales)
print("Average Sales:", average_sales)
```

```python
#Question Three
#Hospital BMI Calculation System


def cal_BMI(weight, height):
    bmi = weight / (height * height)

    if bmi < 18.5:
        category = "UNDERWEIGHT"
    elif bmi < 25:
        category = "NORMAL"
    elif bmi < 30:
        category = "OVERWEIGHT"
    else:
        category = "OBESE"

    return bmi, category



weight = float(input("Enter weight in kilograms: "))
height = float(input("Enter height in meters: "))

BMI, category = cal_BMI(weight, height)

print("BMI:", round(BMI, 2))
print("Category:", category)

```
