1. The population of a town is 198568. Out of them 45312 are men and 35678 are women. Find the number of children in the town.

```python
#Number of Children
total_population = 198568
men = 45312
women = 35678
children = total_population - (men + women)
print("1. Number of children:", children)
```

2. A shopkeeper has 2425 boxes of 24 pencils each. How many pencils do all the boxes have in all?

```python
# 2. Boxes of pencils
boxes = 2425
pencils_per_box = 24
total_pencils = boxes * pencils_per_box
print("2. Total pencils:", total_pencils)

```

3. Linda bought a graduation suit for CFA 45000 and a pair of shoe for CFA 60000. She gave $5000 to the shopkeeper. How much money did the shopkeeper return to her?

4. The cost of 21 TV sets is CFA 19785000. Find the cost of one TV set.

```python

# 4. Cost of one TV
total_tv_cost = 19785000
number_of_tvs = 21
cost_per_tv = total_tv_cost // number_of_tvs
print("4. Cost of one TV:", cost_per_tv)
```

5. A factory produces 24532 bulbs in a month. What is its annual production?

6. There are 145968 bags of sugar, 236487 bags of wheat and some bags of rice in a store room. If the total number of bags in the store room is 450000, find the number of bags of rice.

```python
# 6. Bags in store room
sugar_bags = 145968
wheat_bags = 236487
total_bags = 450000
rice_bags = total_bags - (sugar_bags + wheat_bags)
print("6. Bags of rice:", rice_bags)
```

7. A factory manufactured 483685 toys in three weeks. The production in first week was 146345 toys and in second week 138152 toys. Find the production in the third week.

```python
# 7. Toy production
total_toys = 483685
week1 = 146345
week2 = 138152
week3 = total_toys - (week1 + week2)
print("7. Third week production:", week3)
```

9. There are 24 classrooms in a LMUI. 986 students study there. what is the Equal number of students sits in each room?

```python
# 9. Students per classroom
students = 986
classrooms = 24
students_per_class = students // classrooms
remainder_students = students % classrooms
print("9. Students per class:", students_per_class, "Remainder:", remainder_students)
```

10. 986 students of LMUI want to go to SEME beach by bus. If one bus can carry 75 students, how many buses are required to carry all the students?

```python
# 10. Buses required
total_students = 986
students_per_bus = 75
buses_required = total_students // students_per_bus
if total_students % students_per_bus != 0:
    buses_required += 1
print("10. Buses required:", buses_required)
```

12. In the recently conducted election, 2474179 people voted for Paul Biya, 1622334 people for Issa Tchiroma and 36824 people for Osih Jushua. If everyone voted in the town, what is the total number of voters?

13. Maria bought 96 toys priced equally for CFA 12960. The amount of CFA 1015 is still left with her. Find the cost of each toy and the amount she had.

```python
# 13. Cost of toys
total_toy_cost = 12960
number_of_toys = 96
money_left = 1015
cost_per_toy = total_toy_cost // number_of_toys
total_money = total_toy_cost + money_left
print("13. Cost per toy:", cost_per_toy)
print("    Total money Maria had:", total_money)
```
